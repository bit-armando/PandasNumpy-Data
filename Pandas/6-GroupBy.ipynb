Permite **agrupar datos en funcion de los demas**

Por ejemplo agrupamos el autor y hacemos un conteo de las demas columnas

```python
import pandas as pd
data = pd.read_csv('bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
data.groupby('Author').count()
```

Agrupar por autor y mostrar las medias

```python
data.groupby('Author').median()
```

Al agrupar el autor pasa a ser indice

Podemos agrupar por autor y hacer la suma de los valores para el indice de **William Davis**

```python
data.groupby('Author').sum().loc['William Davis']
```

La funciÃ³n `.agg()` permite aplicar varias funciones al DataFrame una vez agrupado segÃºn una columna especÃ­fica. Agrupar por Author y mostrar el mÃ­nimo y mÃ¡ximo de las demÃ¡s columnas

```python
data.groupby('Author').agg(['min','max'])
```

Agrupar por Author, obtener el mÃ­nimo y mÃ¡ximo de la columna â€˜Reviewsâ€™ y sumar los valores de la columna â€˜User Ratingâ€™

```python
data.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```
